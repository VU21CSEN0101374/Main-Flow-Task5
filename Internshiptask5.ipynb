```python
 questions = ["1. How many people have heart disease and how many people do not have heart disease?",
              "2. Which sex has the highest prevalence of heart disease?",
              "3. Which type of chest pain is most common in each sex?",
              "4. Which type of chest pain is most likely to be associated with heart disease?"]

questions
```

```python
#1. How many people have heart disease and how many people do not have heart disease?

df.target.value_counts()
```

```python
#plotting a bar chart
df.target.value_counts().plot(kind = 'bar' , color=["teal","coral"])
plt.title("Heart Disease values")
plt.xlabel("1 = Heart Disease, 0 = No Heart Disease")
plt.ylabel("Amount");
```

```python
#plotting a pie chart
df.target.value_counts().plot(kind = 'pie' , figsize=(7,5))
plt.legend(["Disease","No Disease"])
```

```python
# '0' represent Female
# '1' represent Male
# '0' represent No Disease
# '1' represent Disease
# Now lets check how many 'Male and 'Female' are in dataset

df.sex.value_counts()
```

```python
#plotting a pie chart
df.sex.value_counts().plot(kind = 'pie' , figsize=(7,5))
plt.title("Male Female ratio")
plt.legend(["Male","Female"])
```

```python
# 2. Which sex has the highest prevalence of heart disease?

pd.crosstab(df.target,df.sex)
```

```python
sns.countplot(x='target',data=df,hue='sex')
plt.title("Heart Disease frequency for sex")
plt.xlabel("0 = No heart Disease , 1 = Heart Disease")
```

```python
# 3. Which type of chest pain is most common in each sex?

df.cp.value_counts()
```

```python
#plotting a bar chart
df.cp.value_counts().plot(kind = 'bar' , color=["teal","coral","violet","lightblue"])
plt.title("Type of chest pain vs count");
```

```python
pd.crosstab(df.sex,df.cp)
```

```python
#plotting a bar chart
df.target.value_counts().plot(kind = 'bar' , color=["teal","coral"])
plt.title("Type of chest pain by sex")
plt.xlabel("0 = Female, 1 = Male")
```

```python
# Which type of chest pain is most likely to be associated with heart disease?

pd.crosstab(df.cp,df.target)
```
